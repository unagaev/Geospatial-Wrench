```python
import json
```

```python
Dalwhinnie = {}
```

```python
directory = 'C:/Users/rost8/Downloads/'
```

```python
f = open('C:/Users/rost8/Downloads' + '/Forrest Hut.json', encoding="utf8")
d = json.load(f)
```

```python
d
```

```python
d['layers'][0]['id'], d['layers'][0]['text'], d['layers'][0]['area'], d['layers'][0]['textLabelOptions']['text']
```

```python
Farm = 'Forrest Hut'
```

```python
coordinates = []

for i in d['layers'][0]['paths']:

    coordinates.append([i['lng'], i['lat']])
coordinates.append([d['layers'][0]['paths'][0]['lng'], d['layers'][0]['paths'][0]['lat']])

Dalwhinnie['type'] = 'FeatureCollection'
inner_feature = {}
inner_feature['type'] = 'Feature'
inner_feature['geometry'] = {"type": "Polygon",
        "coordinates": [coordinates]}
inner_feature['properties'] = {'id': d['layers'][0]['id'],  'Farm': Farm, 'Block': d['layers'][0]['text'], 'Area': d['layers'][0]['area'], 'Area_ha': d['layers'][0]['textLabelOptions']['text'].split('┘')[1]}
Dalwhinnie['features'] = []
Dalwhinnie['features'].append(inner_feature)
```

```python
coordinates = []

for i in d['layers'][1]['paths']:

    coordinates.append([i['lng'], i['lat']])
coordinates.append([d['layers'][1]['paths'][0]['lng'], d['layers'][1]['paths'][0]['lat']])
inner_feature = {}
inner_feature['type'] = 'Feature'
inner_feature['geometry'] = {"type": "Polygon",
        "coordinates": [coordinates]}
inner_feature['properties'] = {'id': d['layers'][1]['id'], 'Farm': Farm, 'Block': d['layers'][1]['text'], 'Area': d['layers'][1]['area'], 'Area_ha': d['layers'][1]['textLabelOptions']['text'].split('┘')[1]}
Dalwhinnie['features'].append(inner_feature)
```

```python
coordinates = []

for i in d['layers'][2]['paths']:

    coordinates.append([i['lng'], i['lat']])
coordinates.append([d['layers'][2]['paths'][0]['lng'], d['layers'][2]['paths'][0]['lat']])
inner_feature = {}
inner_feature['type'] = 'Feature'
inner_feature['geometry'] = {"type": "Polygon",
        "coordinates": [coordinates]}
inner_feature['properties'] = {'id': d['layers'][2]['id'], 'Farm': Farm, 'Block': d['layers'][2]['text'], 'Area': d['layers'][2]['area'], 'Area_ha': d['layers'][2]['textLabelOptions']['text'].split('┘')[1]}
Dalwhinnie['features'].append(inner_feature)
```

```python
with open(directory+f'{Farm}1.json', 'w', encoding='utf-8') as md:
    json.dump(Dalwhinnie, md, ensure_ascii=False, indent=4)
```
